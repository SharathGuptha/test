```python
y=Titanic_Data["Survived"]
X=Titanic_Data.drop('Survived',axis=1)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2,random_state=123)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
num_attr=X_train.select_dtypes(['int64','float64']).columns
num_attr
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train[num_attr]=scaler.fit_transform(X_train[num_attr])
X_test[num_attr]=scaler.fit_transform(X_test[num_attr])
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(X_train,y_train)
X_train_pred=LR.predict(X_train)
X_test_pred=LR.predict(X_test)
```

```python
LR
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
from sklearn.metrics import confusion_matrix
```

```python
print(accuracy_score(y_train,X_train_pred))
print(recall_score(y_train,X_train_pred))
print(f1_score(y_train,X_train_pred))
```

```python
print(accuracy_score(y_test,X_test_pred))
print(recall_score(y_test,X_test_pred))
print(f1_score(y_test,X_test_pred))
```

```python
print('========Train=======')
print(f"Confusion Matrix \n{confusion_matrix(y_train, X_train_pred)}")
print('========Test=======')
print(f"Confusion Matrix \n{confusion_matrix(y_test, X_test_pred)}")
```

```python
Q1,Q3 = np.percentile(Titanic_Data["Age"], [25,75])
```

```python
IQR = Q3 - Q1
ul = Q3+1.5*IQR
ll = Q1-1.5*IQR
```

```python
Data_wo = Titanic_Data[(Titanic_Data["Age"]>ll)&(Titanic_Data["Age"]<ul)] #dropping outliers
```

```python
X_wo = Data_wo.loc[:,Data_wo.columns.difference(['Survived'])] # taking all the independent columns
y_wo = Data_wo.Survived # separting the target variable
```

```python
X_train_wo, X_test_wo, y_train_wo, y_test_wo = train_test_split(X_wo, y_wo, stratify= y_wo, test_size = 0.2, random_state=2323)

print(X_train_wo.shape)
print(y_train_wo.shape)
print(X_test_wo.shape)
print(y_test_wo.shape)
```

```python
cat_attr=X_train_wo.select_dtypes(include ='object').columns #seperating categorucal and numerical attributes
num_attr=X_train_wo.select_dtypes(include =['int64','float64']).columns 
```

```python
wo_model = LogisticRegression().fit(X_train_wo,y_train_wo) #buidlimg model

y_train_Pred_wo = wo_model.predict(X_train_wo)
y_test_Pred_wo = wo_model.predict(X_test_wo)
```

```python
print('========Train=======')#confusion matrix of train dfs
print(f"Confusion Matrix \n{confusion_matrix(y_train_wo, y_train_Pred_wo)}")
print('========Test=======')#confusion matrix of test dfs
print(f"Confusion Matrix \n{confusion_matrix(y_test_wo, y_test_Pred_wo)}")
```

```python
print(accuracy_score(y_train_wo,y_train_Pred_wo))
print(recall_score(y_train_wo,y_train_Pred_wo))
print(f1_score(y_train_wo,y_train_Pred_wo))
```

```python
print(accuracy_score(y_test_wo,y_test_Pred_wo))
print(recall_score(y_test_wo,y_test_Pred_wo))
print(f1_score(y_test_wo,y_test_Pred_wo))
```

```python
from mlxtend.plotting import plot_learning_curves
plot_learning_curves(X_train_wo, y_train_wo, X_test_wo, y_test_wo,wo_model,scoring="f1")
plt.show()
```
